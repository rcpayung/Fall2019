# What is a Dashboard?

"A dashboard is a visual display of the most important information needed to achieve one or more objectives that has consolidated on a single computer screen so that it can be monitored at a glance" -Stephan Few (2006) 
  
<img src="./images/few-0.png" alt="Drawing" align="left" style="width: 200px" />

# Definition of an Information Dashboard?

The definition is evolving from single purpose/single view displays to multiple interactive views with a more complex set of objectives including communication, learning and motivation as well as the traditional purposes of monitoring and decision support.   

* **A Dashboard is:**
    - Visual - combines multiple individual visualizations (bar plots, lineplots, pie charts, histograms, gauges, etc...) in a tiled format
        - Each of which should be built based on the principles of good data visualization design
    - Interrelated - the individual visualizations are often interelated with each other in some way
    - Interactive - allowing users to interact with explore and analyze the data in some way
    - Functional - focused on specific objectves and showing critical information at a glance
        - Sometimes becoming a central location for interested parties to monitor and analyze performance
    - Timely (sometimes) - allowing users to know what's happening now  
* **Key Applications of Dashboards:**
    - Monitoring
    - Decision Support
    - Monitoring of Key Performance Indicators (KPIs) of a business or process
    - Situational Awareness

# Python Visualization - Introduction

```python
import os
import numpy as np
import pandas as pd
from pandas.plotting import register_matplotlib_converters
import matplotlib.pyplot as plt
import seaborn as sns
register_matplotlib_converters()
```

```python
# check working directory
cwd = os.getcwd()
print(cwd)
```

```python
# read the data using pandas into a dataframe - note this expects the data to be in the current working directory 
df = pd.read_csv('./data/rps-0.csv')

# seaborn is built as a set of methods on top of matplotlib - this is a matplotlib function to set the figure size and resolution
f = plt.figure(figsize=(8, 6),dpi=100)

# set the background grid style - seaborn has a range of styles - this one removes the grid and just sets tick marks
sns.set(style='ticks')

# seaborn as support for pandas dataframes - here I'm just selecting the columns I want in the plot in the order I want them
df = df[ ['n-mix', 'n-opp', 'n-rock', 'n-paper', 'n-scissors']]

# use a lineplot - returns matplotlib axes object with plot drawn into it
ax = sns.lineplot(data=df, palette='colorblind', dashes=False, linewidth=0.8)

# set title,  legend, and axes labels
ax.set_title('Rock-Paper-Scissors ABM Output')
ax.set_xlabel('round')
ax.set_ylabel('number of agents')
ax.legend(['Mix', 'Opponent', 'Rock', 'Paper', 'Scissors'])

# save the results
plt.savefig('./results/fig1.png', dpi=100)
```
