u = 20

v = 30


```python
G = netx.Graph();
```

```python
for u in range(1,21):
    for v in range(1,31):
        G.add_node((u,v));
        
```

```python
for u in range(1,21):
    for v in range(1,31):
        G.add_edge((u,v),(u+1,v));
        G.add_edge((u,v),(u,v+1));
        if (u == v):
            if (G.has_node((u,v)) and G.has_node((u+1,v+1))):
                G.add_edge((u,v),(u+1,v+1));
```

```python
options = {'node_size':50,'width':0.2};
plt.figure(figsize=(5,5));
netx.draw_random(G, **options)
```

```python
h = {};

for u in range(1,21):
    for v in range(1,31):
        if (G.degree((u,v)) in h):
            h[G.degree((u,v))] += 1;
        else:
            h[G.degree((u,v))] = 1;
            
```

```python
print(h)
```

```python
plt.bar(h.keys(),h.values());
plt.xticks(np.arange(1,7,1))
plt.title("Degree Histogram");
plt.xlabel("Degree");
plt.ylabel("# of Nodes of Degree K")
plt.show()
```
