```python
# Create DF2:

df2 = pd.read_sql_query(
'''
    SELECT Event.e_id, COUNT(User.u_id) as 'Users Attended'
    FROM User NATURAL JOIN Event, Participates
    WHERE User.u_id = Participates.u_id AND Event.e_id = Participates.e_id
    GROUP BY Event.e_id

''',conn);
df2
```

```python
# Create the harvesine distance function.
from math import sin, cos, sqrt, atan2, radians
def harvesine_dist(lat1, lon1, lat2, lon2):
    
    lat1 = radians(lat1);
    lat2 = radians(lat2);
    lon1 = radians(lon1);
    lon2 = radians(lon2);
    
    R = 6373.0 # approximate radius of earth in km

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# calculate the distances and add a new data column to df1 (Distance (KM))
distance = [];
print(df1.dtypes)
for i in df1.index:
    distance.append(harvesine_dist(float(df1['EVENT Lat'][i]),float(df1['EVENT Lon'][i]),float(df1['USER Lat'][i]),float(df1['USER Lon'][i])));
df1['distance (KM)'] = distance;
df1
```
