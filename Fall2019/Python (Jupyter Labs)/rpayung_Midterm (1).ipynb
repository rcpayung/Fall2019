```python
def price_inflation(rate,year,currentprice):
    for n in range(year):
        currentprice = currentprice * (1 + (rate/100));
    return currentprice;

print(price_inflation(5,1,100))
print(price_inflation(5,2,100))
print(price_inflation(0.5,15,10))
```

## Question 4

```python
def projectile_launch(velocity,theta):
    g = 9.81;
    T = (2 * velocity * math.sin(math.radians(theta))) / g;
    R = (math.pow(velocity,2) * math.sin(2*math.radians(theta))) / g;
    Hmax = (math.pow(velocity,2) * math.pow(math.sin(math.radians(theta)), 2)) / (2*g);
    
    return T,R,Hmax

print(projectile_launch(10,30))
print(projectile_launch(10,45))
print(projectile_launch(10,60))
```

## Question 6

```python
def findTime(vx, distance):
    return distance / vx;
# returns the time taken to travel distance.

def distance(x,y):
    return math.sqrt(x**2 + y++2);
# returns the distance between two objects

def findHeight(t):
    vy = 0;
    y2 = 0;
    g = 9.8;
    
    y1 = vy * t + (1/2)*g*(t**2) - y2;
    return y1;
# returns the height required given a specified value of time t. 

#totaltime takes the velocity along the x axis and total distance to travel (I took the values from the green object).
totaltime = findTime(3,6);

# finds the height required for the two objects to collide at the same time using totaltime.
heightRequired = findHeight(totaltime);

def green(vx,x1,algorithm=1):
    greenList = []

    for n in range(1,int(algorithm*totaltime)):
        t = n/algorithm
        x2 = x1 + vx * t
        greenList.append(x2);
    return greenList;
# calculates the list of positions in time based on the initial velocity x, final position, and algorithm which allows us to edit the number of steps taken per measurement.
# the default will print the initial position of the object. Specifying 10 for algorithm will allow us to take 0.1 second steps.

def orange(y1, algorithm=1):
    g = -9.8;
    vy = 0;
    positionlist = []
    for n in range(1,int(algorithm * totaltime)):
        t = n/algorithm;
        y2 = y1 + t*vy + ((1/2) * g * t**2)
        positionlist.append(y2)
    return positionlist
# calculates the list of positions in time based on an initial height and algorithm which allows us to edit the number of steps taken per measurement.
# the default will print the initial position of the object. Specifying 10 for algorithm will allow us to take 0.1 second steps.

greenList = green(3,0,10);
orangeList = orange(heightRequired,10);

print("Green Position List: ", greenList);
print();
print("Orange Position List: ", orangeList);

listfinal = [];
# Calculates the distance between the objects, goes backwards in the greenList since the green list is increasing while the orange list decreases.
for n in range(len(orangeList)):
    listfinal.append(distance(orangeList[n],greenList[-n]))

print()
print("Distance between objects List: ",listfinal)
plt.plot(listfinal);
plt.xlabel("Observation #");
plt.ylabel("Distance")
plt.show()
```
