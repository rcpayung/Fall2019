```python
def getData(ticker ='AAPL',datefrom='2018-01-01',dateto=date.today().strftime('%Y-%m-%d')):
    api_result = rq.get('https://api.marketstack.com/v1/tickers/' + ticker + '/eod', {'access_key':'3bda58441e22216933041da59f79c812', 'date_from':datefrom, 'date_to':dateto, 'limit':1000});
    response = api_result.json();
    df = pd.DataFrame.from_dict(response['data']['eod']);
    dates = [];
    for i in df['date']:
        dates.append(i[0:10])
    df['date'] = pd.DatetimeIndex(dates)
    df.set_index('date',inplace=True)
    df = df.reindex(index=df.index[::-1])
    return df;
```

```python
dfAAPL = getData()
```

```python
dfAAPL['close'].plot(figsize=(20,10))
```

```python
def fill_nan(df):
    NANrows = pd.isnull(df).any(1).nonzero()[0]
    null_df = df.isnull()
    for row in NANrows :
        for colum in range(0,df.shape[1]):
            if(null_df.iloc[row][colum]):
                df.iloc[row][colum] = (df.iloc[row-1][colum]+df.iloc[row-1][colum])/2
    return df
```

```python
# Resampled over business days.
# Rolling average
dfAAPL2 = dfAAPL.resample('b').mean()
dfAAPL2 = fill_nan(dfAAPL2)
dfAAPL2.rolling(window=10).mean()['close'].plot(figsize=(20,10),lw=4,color='k')
dfAAPL['close'].plot(lw=1,color='b')
```

```python
dfAAPL2['EWMA'] = dfAAPL2['close'].ewm(span=6).mean()
```

```python
# Exponentially weighted moving average
dfAAPL2['EWMA'].plot(lw=3,figsize=(20,10))
dfAAPL2['close'].plot(lw=1)
plt.xlabel("Date")
plt.ylabel("Stock Price")
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
fit_model = ExponentialSmoothing(dfAAPL2['close'],trend='add',seasonal='add',seasonal_periods=180).fit()
```

```python
dfAAPL2['des'] = fit_model.fittedvalues.shift(-1)
```

```python
dfAAPL2[['close','des']].plot(figsize=(20,10))
```

```python
dfAAPL2[['close','des']].iloc[-200:].plot(figsize=(20,10))
```
