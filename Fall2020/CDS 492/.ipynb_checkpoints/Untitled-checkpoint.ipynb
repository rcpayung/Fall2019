```python
def requestLoader(ticker = "AAPL",requested = 'eod'):
    api_result = requests.get('https://api.marketstack.com/v1/tickers/' + ticker + '/' + requested, {'access_key':'3bda58441e22216933041da59f79c812'});
    response = api_result.json();
    return response;

def responseFrame(response, param = 'eod'):
    df = pd.DataFrame.from_dict(response['data'][param]);
    return df;
```

```python
top100 = ["BABA","GOOGL","AMZN","AMC","AMD","AAL","APHA","AAPL","AZN","T","ACB","BAC","BYND","BA","BP","CGC","CCL","CPRX","KO","CRON","PLAY","DAL","GUSH","DIS","ET","XOM","FB","FIT","F","FCEL","GE","GILD"]
```

```python
dfs = [];
for i in top100:
    print(i)
    dfs.append(responseFrame(requestLoader(ticker = i)))
```

```python
def searchForSplits(df,threshold = 1.5): #data frame, threshold = more than halved.
    count = 0;
    for i in range(len(df)-1):
        if (abs(1 - (df['close'][i+1] / df['close'][i])) > 1):
            print("A possible split occurred for %s between: %s and %s" % (df['symbol'][0],df['date'][i+1],df['date'][i]))
            count += 1;
    if (count == 0):
        print("No Stock Split occurred within the given time frames for the stock: ", df['symbol'][0])
            
```

```python
dfTSLA = responseFrame(requestLoader(ticker="TSLA"));
searchForSplits(dfTSLA)
print(abs(1-(dfTSLA['close'][22] / dfTSLA['open'][21]))+1)
```

```python
def dfNormalize(df):
    dfNormal = [];
    for i in range(len(dfTSLA)-1):
        dfNormal.append(abs(df['close'][i+1] / df['open'][i])-1);
    return dfNormal;
```

```python
plt.plot(dfNormalize(dfTSLA),linewidth=6,label="TSLA",color='k')
plt.plot(dfNormalize(dfs[7]),linewidth=3,label="AAPL",color='m')
plt.plot(dfNormalize(dfs[6]),linewidth=2,label=dfs[3]['symbol'][0],color='red')
plt.legend()
plt.savefig("Plot generated using Matplotlib.png")
plt.show()
```

```python

import numpy as np
dfNormalizedTSLA = dfNormalize(dfTSLA)
```
