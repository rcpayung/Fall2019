## Problem 3

```python
# your code an a plot
dt = 0.1;
k = 10;
m = 2;
dx = 0.3;
x = dx;
v = (k/(0.5*m))**(0.5)
a = -k*dx/m;
pos = [];
for i in range(1000):
    if (x > -0.05 and x < 0.05):
        v = v * 0.9
    a2 = -k*x/m;
    v2 = v + a2*dt*0.5
    x2 = x + v * dt + 0.5 * a2 * dt * dt
    v = v2;
    x = x2;
    pos.append(x2);
plt.plot(pos)
plt.show()
```
