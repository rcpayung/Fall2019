```python
import numpy as np
import matplotlib.pyplot as plt
import floyd
import bacon
import pandas as pd
import movies3 as mvs
```

```python
movies, actors, isin = mvs.ReadData("movies1000.xlsx")
```

```python
def moviesByYear(year,movies):
    moviesResult = []
    for i in movies:
        if (i[2]==year):
            moviesResult.append(i[0]);
    return moviesResult;
```

```python
movies2007 = moviesByYear(2007,movies)
```

```python
aids = mvs.AidsFromMids(isin,movies2007)
```

```python
def MakeG(movies, isin):
    mat = np.array(movies);
    t = mat[:,0]+0
    mids = np.array(list(set(t)))
    N = len(mids)
    G = np.zeros((N+1,N+1))

    for i in mids:
        aids = mvs.AidsFromMid(isin, i);
        mids2 = mvs.MidsFromAids(isin, aids);
        for j in mids2:
            G[i-1,j-1] = 1;
    return G;
```

```python
G = MakeG(movies,isin)
```

```python
g,p = bacon.RunFloyd(G)
```

```python
def YACL(mids):
    AverageChainLength = [];
    count = 0;
    pathlength = 0;
    for j in mids:
        for i in movies:
            pathlength += len(floyd.FindPath(p,j,i[0]))
            count += 1;
        AverageChainLength.append(pathlength/count);
        pathlength = 0;
        count = 0;
    return AverageChainLength;
```

```python
yacl = YACL(movies2007)
```

```python
plt.scatter(movies2007,yacl)
plt.show()
```

```python
np.mean(yacl) # 2007's YACL
```

```python
def Run(yearLow, yearHigh):
    midsYears = [];
    yaclYears = [];
    for i in range(yearLow,yearHigh):
        mids = moviesByYear(i,movies);
        midsYears.append(mids);
        yaclYears.append(YACL(mids));
    return midsYears, yaclYears
```

```python
def Build(yearLow, yearHigh):
    mov, yacls = Run(yearLow,yearHigh);

    plt.figure(figsize=(15,8))
    for i in range(yearHigh-yearLow):
        plt.scatter(mov[i],yacls[i],marker='o',label=str(i+yearLow))
    plt.legend()
    plt.xlabel("Mids");
    plt.ylabel("YACL");
    plt.title("YACL vs Mids from "+str(yearLow)+" to "+str(yearHigh-1));
    plt.show()

    yaclAvg = 0;
    for i in yacls:
        if (len(i) > 0):
            yaclAvg += (sum(i)/len(i))
    yaclAvg = yaclAvg/len(yacls)
    return yaclAvg;
```

```python
years = [1950,1960,1970,1980,1990,2000,2010]
yacls = [];
for i in years:
    yacls.append(Build(i,i+10));
```

```python
plt.plot(years,yacls)
plt.xlabel("Decade");
plt.ylabel("YACL");
plt.title("Average YACL by decades since 1950")
plt.show()
```
